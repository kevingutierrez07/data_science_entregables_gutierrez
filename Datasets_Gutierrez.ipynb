```python
# Conexi√≥n al dataset dividido
set_part_1 = pd.read_csv('https://raw.githubusercontent.com/kevingutierrez07/data_science_entregables_gutierrez/main/application_record_1.csv')
set_part_2 = pd.read_csv('https://raw.githubusercontent.com/kevingutierrez07/data_science_entregables_gutierrez/main/application_record_2.csv')
set_part_3 = pd.read_csv('https://raw.githubusercontent.com/kevingutierrez07/data_science_entregables_gutierrez/main/application_record_3.csv')
set_part_4 = pd.read_csv('https://raw.githubusercontent.com/kevingutierrez07/data_science_entregables_gutierrez/main/application_record_4.csv')
set_credit = pd.read_csv('https://raw.githubusercontent.com/kevingutierrez07/data_science_entregables_gutierrez/main/credit_record.csv')

# Realizar las uniones utilizando la columna 'ID' como clave
set_completo = pd.merge(set_part_1, set_part_2, on='ID', how='outer')
set_completo = pd.merge(set_completo, set_part_3, on='ID', how='outer')
set_completo = pd.merge(set_completo, set_part_4, on='ID', how='outer')

# Verificar el resultado
print(set_completo.head(10))
```
